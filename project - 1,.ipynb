```python
# Q4.find out the outliers for "Total bill" and "tip"
# Q5. Find out the IQR
```

```python
from numpy import percentile
```

```python
quartiles = percentile(df["total_bill"], [25,50,75])
dmin,dmax = df["total_bill"].min(),df["total_bill"].max()
```

```python
print("Min :", dmin)
print("Quartile Q1 :", quartiles[0])
print("Quartile Q2:", quartiles[1])
print("Quartile Q3:", quartiles[2])
print("Max :", dmax)
```

```python
iqr = quartiles[2]- quartiles[0]
print("Interquartile range:", iqr)
```

```python
outlmin = quartiles[0]- (1.5*iqr)
outlmax = quartiles[2] +(1.5*iqr)
print("Minimun outliers =", outlmin)
print("Maximun outliers =", outlmax)
```

```python
quartiles = percentile(df["tip"], [25,50,75])
dmin,dmax = df["tip"].min(),df["tip"].max()
```

```python
print("Min :", dmin)
print("Quartile Q1 :", quartiles[0])
print("Quartile Q2:", quartiles[1])
print("Quartile Q3:", quartiles[2])
print("Max :", dmax)
iqr = quartiles[2]- quartiles[0]
print("Interquartile range:", iqr)
```

```python
outlmin = quartiles[0]- (1.5*iqr)
outlmax = quartiles[2] +(1.5*iqr)
print("Minimun outliers =", outlmin)
print("Maximun outliers =", outlmax)
```

```python
# Q8. Plot the cumulative frequency polygon for "total_bill" and "tip"
```

```python
values,base = np.histogram(df ["tip"])
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative,c = 'r')
plt.title('Tip')
plt.show()
```

```python
values,base = np.histogram(df ["total_bill"])
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative,c = 'r')
plt.title('Total_bill')
plt.show()
```
